**LOADING DATA**

```python
import pandas as pd

iris = pd.read_excel('iris_dataset.xlsx', sheet_name='iris')
iris.head()
```

**PREPROCESSING**
* Removing unnecessary columns
* Removing duplicates if it exist

_when we saved the data in an excel file we forgot to deactivate the auto indexing_

```python
iris.drop(columns='Unnamed: 0', inplace=True)
iris.drop_duplicates(inplace=True)
```

**EXPLORATORY**
* As we can see, the dataset is devided into 3 classes under the names ***setosa***, ***versicolor*** and ***virginica***
* 

```python
import seaborn as sns

g = sns.pairplot(iris, hue="species", palette="Paired", diag_kind="hist", height=2.5)
```
